selecting the numbers of variables

```python
data=df[["Area","Price"]]
```

```python
import matplotlib.pyplot as plt

data.plot(x="Area",y="Price", style='o')
plt.xlabel("AREA")
plt.ylabel("PRICE")
plt.show()
```

seperate the dependent and independent(all) variable...

```python
x=pd.DataFrame(data["Area"])
```

```python
y=pd.DataFrame(data["Price"])
```

seperate the train and test datasets ..


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

printing the shape of x_train,x_test,y_test,y_train

```python
print(f"the shape of x_train :{x_train.shape}")
print(f"the shape of y_train:{y_train.shape}")

print("x"*30)
print(f"the shape of x_test :{x_test.shape}")
print(f"the shape of y_test:{y_test.shape}")
```

```python
# prompt: importing the linear regression model

from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

fitting the  model in x_train and y_train

```python
# Select only numerical columns for training
model.fit(x_train,y_train)
```

printing the coef(m) ---> tht is slope of y=mx+c line of linear regression

```python
print(model.coef_)
```

printing the y-intercept(C) of the line y=mx+c  of linear regression

```python
print(model.intercept_)
```

predicting the value of y_test on the basis of x_test...

```python
y_pred=model.predict(x_test)
y_pred=pd.DataFrame(y_pred,columns=["predicted value of y"])
y_pred
```

```python
y_pred
```

```python
y_test
```

calculating the errors....

```python
from sklearn import metrics

print(f"MAE:{metrics.mean_absolute_error(y_test,y_pred)}")
print(f"MSE:{metrics.mean_squared_error(y_test,y_pred)}")
print(f"RMSE:{np.sqrt(metrics.mean_squared_error(y_test,y_pred))}")
print(f"R squared :{metrics.r2_score(y_test,y_pred)}")

```
